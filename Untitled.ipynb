```python
np.random.seed(1)
weights=np.random.rand(5,1)
lr, bias = 0.05,0.3
```

```python
weights
```

```python
def sigmoid(val):
    
    return 1/(1+np.exp(-val)) 
```

```python
def sigmoid_der(val):
    return sigmoid(val)*(1-sigmoid(val))
```

```python

for i in range(1000):
    inputs = X
    input_1 = np.dot(X,weights)+bias
    
    z = sigmoid(input_1)
    
    error = (z - y)
    
    print(error.sum())
    
    temp_cost = error
    temp_pred = sigmoid_der(z)
    prod = temp_cost * temp_pred
    
    weights = weights - lr*np.dot(inputs.T, prod)
    
    for num in prod:
        bias = bias - lr*num
```
